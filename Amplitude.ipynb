### Solving the Amplitude with a general equation and plotting the amplitude over frequencies

```python
import numpy as np
import matplotlib.pyplot as plt
# This makes the plots appear inside the notebook
%matplotlib inline
from scipy.integrate import simps
from numpy import trapz



# Constants
m = 6246.00
L = 5.48
c = 10^4
g = 9.81


k = np.linspace(2,7,6)*10**10
b = np.linspace(9000,12000,300)*2*np.pi/60


Amplitude = np.zeros((len(k),len(b)))
```

```python
T0 = 100

def calcAmp(T0):
    for i in range(len(k)):
        A = T0/(-m*L**2*b**2+(c**2*b**2)/(m*g*L+k[i]*L-m*L**2*b**2)+m*g*L+k[i]*L)
        B = -c*b/(m*g*L+k[i]*L-m*L**2*b**2)*A
        Amplitude[i,:] = np.sqrt(A**2+B**2)
    return Amplitude


calcAmp(T0)
```

```python
def plotAmp(Amplitude,T0):
    plt.figure(figsize=(10,6))

    plt.title("Amplitude vs. Frequency of the Engine Vibration with initial Torque of "+str(T0)+" N")
    plt.xlabel("Amplitude of Steady State Oscillation")
    plt.ylabel("Frequency of Vibration [rad/s]")
    for i in range(len(k)):
        plt.plot(b,Amplitude[i,:], label = "K value "+str(k[i]))
    
    plt.legend()
    plt.show()
    
    
plotAmp(Amplitude,T0)
```
